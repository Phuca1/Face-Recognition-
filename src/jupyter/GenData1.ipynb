```
import time
import dlib
import cv2
import os

import glob
import random
```

```
def detection(image):
  hog_face_detector = dlib.get_frontal_face_detector()
  faces = hog_face_detector(image, 1)
  if not faces:
    return None
  face = faces[0]
  return image[(face.top()-10):(face.bottom()+10), (face.left()-10):(face.right()+10), :]
```

```
def extract_img(video_path, out_dir_train, out_dir_test):
    videos = glob.glob(os.path.join(video_path, "*.*"))
    folder_name = os.path.basename(video_path).split('.')[0]
    result_path_train = os.path.join(out_dir_train, folder_name)
    result_path_test = os.path.join(out_dir_test, folder_name)
    os.makedirs(result_path_train, exist_ok=True)
    os.makedirs(result_path_test, exist_ok=True)
    count = 0
    for video_path in videos:
        video = cv2.VideoCapture(video_path)
        print(video)
        success, frame = video.read()
        while success:
            if count == 2000:
              break
            if count % 10 == 0:
              try:
                  face = detection(frame)
                  if face is None:
                      success, frame = video.read()
                      continue
                
                  img = cv2.resize(face, (224,224))
                  cv2.imwrite('{}/{}.jpg'.format(result_path_train, random.randint(0,1000000)), img)
              except Exception:
                print('some thing wrong')
            elif count % 37 == 0:
              try:
                  face = detection(frame)
                  if face is None:
                      success, frame = video.read()
                      continue
                
                  img = cv2.resize(face, (224,224))
                  cv2.imwrite('{}/{}.jpg'.format(result_path_test, random.randint(0,10000000)), img)
              except Exception:
                print('some thing wrong')
            count+=1
            success, frame = video.read()
```

```
videos_dir = "/content/drive/MyDrive/TrainImage/video/"
image_path_train = "/content/drive/MyDrive/TrainImage/image/train/"
image_path_test = "/content/drive/MyDrive/TrainImage/image/test/"
list_videos_folder = glob.glob(os.path.join(videos_dir, "*"))
for item in list_videos_folder:
  extract_img(item, image_path_train,image_path_test)
```
